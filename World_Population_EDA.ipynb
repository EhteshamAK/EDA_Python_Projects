```python
# change column names
rename_cols = ({
    'Rank':'rank',
    'CCA3': 'country_short3',
    'Country': 'country',
    'Capital': 'capital',
    'Continent': 'continent',
    '2022 Population': '2022_pop',
    '2020 Population': '2020_pop',
    '2015 Population': '2015_pop',
    '2010 Population': '2010_pop',
    '2000 Population': '2000_pop',
    '1990 Population': '1990_pop',
    '1980 Population': '1980_pop',
    '1970 Population': '1970_pop',
    'Area (km2)': 'Area_km2',
    'Density (per km2)': 'density_per_km2',
    'Growth Rate': 'growth_rate',
    'World Population Percentage': 'world_pop_percentage',
 })

df.rename(rename_cols, axis = 1, inplace = True)
```

## Fill the missing values in columns using interpolate

```python
# calculae the interpolate for 

fill_missing_values_in_columns = {
 '2022_pop':df['2022_pop'].interpolate(),
    '2022_pop':df['2022_pop'].interpolate(),
    '2020_pop':df['2020_pop'].interpolate(),
    '2015_pop':df['2015_pop'].interpolate(),
    '2010_pop':df['2010_pop'].interpolate(),
    '2000_pop':df['2000_pop'].interpolate(),
    '1990_pop':df['1990_pop'].interpolate(),
    '1980_pop':df['1980_pop'].interpolate(),
    '1970_pop':df['1970_pop'].interpolate(),
    'growth_rate':df['growth_rate'].interpolate(),
}

df = df.fillna(fill_missing_values_in_columns)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df = df.rename(columns=({'Area (km²)': 'area_km2','Density (per km²)':'density_per_km2'}))
```

```python
df.columns
```

```python
missing_val = {
    'area_km2':df['area_km2'].interpolate(),
    'density_per_km2':df['density_per_km2'].interpolate(),
}

df = df.fillna(missing_val)
```

```python
df.isnull().sum()
```
